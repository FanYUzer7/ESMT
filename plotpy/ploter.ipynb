```python
def read_data_from_csv(path: str) -> pd.DataFrame:
    construct_test_res = pd.read_csv(path)
    a = construct_test_res[['mrt-uni', 'mrt-imis', 'esmt-uni', 'esmt-imis']]
    a.index = construct_test_res['data size'].array / 1000
    return a

def draw_plot(data: pd.DataFrame, title: str):
    with plt.style.context(['science', 'ieee', 'no-latex', 'cjk-sc-font']):
        f, ax = plt.subplots()
        sns.lineplot(data=data, markers=['^', 's', 'o', '*'])
        ax.set_xlabel("data size(kelems)")
        ax.set_xscale(value='log', base=2)
        ax.set_ylabel('time(ms)')
        ax.set_title(title)
```

```python
draw_plot(data=read_data_from_csv('./csv/construct.csv'), title='construct')
draw_plot(data=read_data_from_csv('./csv/insert.csv'), title='insert')
draw_plot(data=read_data_from_csv('./csv/delete.csv'), title='delete')
draw_plot(data=read_data_from_csv('./csv/update.csv'), title='update')
draw_plot(data=read_data_from_csv('./csv/query.csv'), title='range query')
```
